```python
#read in a xls file exclude the header and footer 
energy_data = pd.read_excel("/Users/kevin/Downloads/AD 450/Energy Indicators.xls", header = None, skipfooter= 32)
energy_data
```

```python
#get rid of the first two columns
energy_data.drop(energy_data.columns[[0,1]], axis = 1, inplace = True)
energy_data
```

```python
#Naming the columns of the dataframe
energy_data.columns = ["Country", "Energy Supply", "Energy Supply per Capita", "% Renewable"]
energy_data.head()
```

```python
import re
#remove parenthesis and/or parenthesis from country names. 
energy_data.Country = energy_data.Country.str.replace(r'[\(\)\d]+','')
```

```python
#renaming several countries
energy_data.Country = energy_data.Country.replace(to_replace ='Republic of Korea', value = 'South Korea', regex = True)
energy_data.Country = energy_data.Country.replace(to_replace ='United States of America', value = 'United States', regex = True) 
energy_data.Country = energy_data.Country.replace(to_replace ='United Kingdom of Great Britain and Northern Ireland', value = 'United Kingdom', regex = True) 
energy_data.Country = energy_data.Country.replace(to_replace ='China, Hong Kong Special Administrative Region', value = 'Hong Kong', regex = True) 
energy_data.Country = energy_data.Country.replace(to_replace ='Bolivia Plurinational State of', value = 'Bolivia', regex = True) 
#
pd.set_option('display.max_rows', len(energy_data.Country))

energy_data.Country
```
